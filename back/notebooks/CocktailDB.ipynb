```python
def str_to_bool(val):
    if pd.isnull(val):  # NaN이면 None으로
        return None
    return str(val).strip().lower() == "true"
```

```python
import pandas as pd

# 1. NaN or float → int or None 처리기
def float_to_int_or_none(val):
    if pd.isnull(val):
        return None
    return int(val)

# 2. 문자열 "True"/"False" 또는 bool → bool or None 처리기
def str_to_bool(val):
    if isinstance(val, bool):
        return val
    if pd.isnull(val):
        return None
    return str(val).strip().lower() == "true"
```

```python
insert_query = text("""
INSERT INTO cocktail_recipe (
    name,
    ingredient1_id, ingredient1_amount,
    ingredient2_id, ingredient2_amount,
    ingredient3_id, ingredient3_amount,
    ingredient4_id, ingredient4_amount,
    garnish_id, method, ice_in_shake, is_on_the_rocks,
    glass_type, abv, summary
)
VALUES (
    :name,
    :ingredient1_id, :ingredient1_amount,
    :ingredient2_id, :ingredient2_amount,
    :ingredient3_id, :ingredient3_amount,
    :ingredient4_id, :ingredient4_amount,
    :garnish_id, :method, :ice_in_shake, :is_on_the_rocks,
    :glass_type, :abv, :summary
)
""")
```

```python
for _, row in df.iterrows():
    connection.execute(insert_query, {
        "name": row["name"],
        "ingredient1_id": float_to_int_or_none(row["ingredient1_id"]),
        "ingredient1_amount": row["ingredient1_amount"],

        "ingredient2_id": float_to_int_or_none(row["ingredient2_id"]),
        "ingredient2_amount": row["ingredient2_amount"] if pd.notnull(row["ingredient2_amount"]) else None,

        "ingredient3_id": float_to_int_or_none(row["ingredient3_id"]),
        "ingredient3_amount": row["ingredient3_amount"] if pd.notnull(row["ingredient3_amount"]) else None,

        "ingredient4_id": float_to_int_or_none(row["ingredient4_id"]),
        "ingredient4_amount": row["ingredient4_amount"] if pd.notnull(row["ingredient4_amount"]) else None,

        "garnish_id": float_to_int_or_none(row["garnish_id"]),

        "method": row["method"],

        "ice_in_shake": str_to_bool(row["ice_in_shake"]),
        "is_on_the_rocks": str_to_bool(row["is_on_the_rocks"]),

        "glass_type": row["glass_type"],
        "abv": int(row["abv"]),
        "summary": row["summary"]
    })
```
