```python
import pandas as pd

# 경로에 맞게 수정
df = pd.read_csv('../data/raw/KakaoData.csv', encoding='cp949')

# 헤더 정리 (첫 줄 제거 후 헤더 재설정)
df = df.iloc[1:].reset_index(drop=True)
df.columns = ["instruction", "output"]

# 결측값 처리 + 문자열 변환
df["instruction"] = df["instruction"].fillna("").astype(str)
df["output"] = df["output"].fillna("").astype(str)

df.head()
```

```python
df["input_length"] = df["instruction"].apply(len)
df["output_length"] = df["output"].apply(len)

df["input_sentences"] = df["instruction"].apply(lambda x: x.count(".") + x.count("?") + x.count("!") + x.count("ㅋ"))
df["output_sentences"] = df["output"].apply(lambda x: x.count(".") + x.count("?") + x.count("!") + x.count("ㅋ"))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm

# 한글 폰트 설정 (Windows 기준)
plt.rcParams['font.family'] = 'Malgun Gothic'  # Mac이면 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df["input_length"], color="skyblue", kde=True, label="입력 길이")
sns.histplot(df["output_length"], color="salmon", kde=True, label="출력 길이")
plt.title("카카오 대화 데이터 입력/출력 길이 분포")
plt.xlabel("문자 수")
plt.ylabel("샘플 수")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df[["input_length", "output_length"]])
plt.title("입력/출력 길이 Boxplot")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x="input_length", y="output_length", data=df, alpha=0.3)
plt.title("입력 길이 vs 출력 길이 관계")
plt.xlabel("입력 길이")
plt.ylabel("출력 길이")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# ✅ 라이브러리 불러오기
import pandas as pd
import json

# ✅ Kakao CSV 불러오기 (경로 수정 필요)
df = pd.read_csv("../data/raw/KakaoData.csv", encoding="cp949")

# ✅ 첫 줄 제거 + 컬럼 이름 지정
df = df.iloc[1:].reset_index(drop=True)
df.columns = ["instruction", "output"]

# ✅ 결측값 처리
df["instruction"] = df["instruction"].fillna("").astype(str)
df["output"] = df["output"].fillna("").astype(str)

# ✅ 메시지 포맷 변환
chat_data = []
for i, row in df.iterrows():
    item = {
        "messages": [
            {"role": "user", "content": row["instruction"]},
            {"role": "assistant", "content": row["output"]}
        ]
    }
    chat_data.append(item)

# ✅ 일부 샘플 확인
import random
random.sample(chat_data, 3)
```

```python
# ✅ .jsonl 형식으로 저장
with open("../data/processed/kakao_chat_format.jsonl", "w", encoding="utf-8") as f:
    for item in chat_data:
        f.write(json.dumps(item, ensure_ascii=False) + "\n")
```
