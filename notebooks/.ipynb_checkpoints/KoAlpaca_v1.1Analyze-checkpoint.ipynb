```python
# 📌 1. 기본 라이브러리 불러오기
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 📌 2. 원본 JSONL 파일 불러오기
file_path = "../data/raw/KoAlpaca_v1.1.jsonl"  # 파일 위치에 따라 수정

data = []
with open(file_path, "r", encoding="utf-8") as f:
    for line in f:
        item = json.loads(line)
        instruction = item.get("instruction", "").strip()
        output = item.get("output", "").strip()
        if instruction and output:
            data.append({
                "instruction": instruction,
                "output": output
            })

df = pd.DataFrame(data)
df.head()
```

```python
# 📌 3. 특징 추출
df["input_length"] = df["instruction"].apply(len)
df["output_length"] = df["output"].apply(len)
df["input_sentences"] = df["instruction"].apply(lambda x: x.count(".") + x.count("!") + x.count("?"))
df["output_sentences"] = df["output"].apply(lambda x: x.count(".") + x.count("!") + x.count("?"))

df.describe()
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

# Windows 기본 폰트로 변경
plt.rcParams['font.family'] = 'Malgun Gothic'  # 또는 'AppleGothic' (Mac)

# 음수 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

# 이후 그래프 다시 실행
```

```python
# 📌 4. 시각화 - 입력/출력 길이 분포
plt.figure(figsize=(12, 6))
sns.histplot(df["input_length"], color="skyblue", kde=True, label="Input Length")
sns.histplot(df["output_length"], color="salmon", kde=True, label="Output Length")
plt.title("Instruction / Output 길이 분포")
plt.xlabel("문자 수")
plt.ylabel("샘플 수")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import json
import os

# 원본 파일 경로 (수정 가능)
file_path = "../data/raw/KoAlpaca_v1.1.jsonl"

# 저장할 디렉토리 및 파일명
output_path = "../data/processed/koalpaca_chat_format.jsonl"
os.makedirs(os.path.dirname(output_path), exist_ok=True)

# 변환 결과를 저장할 리스트
chat_data = []

# 파일 읽고 변환
with open(file_path, "r", encoding="utf-8") as f:
    for line in f:
        item = json.loads(line)
        instruction = item.get("instruction", "").strip()
        output = item.get("output", "").strip()

        if not instruction or not output:
            continue

        # instruction만 사용해서 user 메시지 생성
        chat_item = {
            "messages": [
                {"role": "user", "content": instruction},
                {"role": "assistant", "content": output}
            ]
        }
        chat_data.append(chat_item)

# 저장 (.jsonl 형식)
with open(output_path, "w", encoding="utf-8") as f:
    for item in chat_data:
        f.write(json.dumps(item, ensure_ascii=False) + "\n")
```
