```python
# 1. 필요한 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2. 데이터 불러오기
df = pd.read_csv('../data/raw/all_drinks.csv')  # 경로는 상황에 맞게 조정

df
```

```python
for col in df.columns:
    print(col)
```

```python
# 1. 제거할 컬럼 정의 (불필요한 메타데이터 + 거의 비어있는 재료/용량 컬럼)
columns_to_drop = [
    'strDrinkThumb',  # 이미지 URL
    'strVideo',       # 레시피 영상
    'dateModified',   # 마지막 수정일
    'strIBA',         # IBA 분류
    'Unnamed: 0',     # 인덱스 잔재 (있을 경우)
    'idDrink',        # 고유 ID

    # 거의 비어있는 재료/용량 컬럼
    *[f'strIngredient{i}' for i in range(10, 16)],
    *[f'strMeasure{i}' for i in range(10, 16)]
]

# 2. 필수 값이 없는 행 제거 (이름, 제조법 기준)
clean_df = df.dropna(subset=['strDrink', 'strInstructions'])

# 3. 지정한 컬럼 제거
clean_df = clean_df.drop(columns=columns_to_drop, errors='ignore')

# 4. 결과 미리 보기
clean_df
```

```python
# 네그로니 칵테일 이름이 포함된 레코드 검색 (대소문자 무시)
negroni_df = clean_df[clean_df['strDrink'].str.contains('negroni', case=False, na=False)]
negroni_df.head()
```

```python
# 1. 찾고자 하는 칵테일 이름 (영문 키워드 기준)
cocktail_keywords = [
    'negroni', 'dry martini', 'last word', 'martinez', 'monkey gland', 'tuxedo', 'hanky panky',
    'aviation', 'bee\'s knees', 'french 75', 'southside', 'white lady', 'tom collins',
    'daiquiri', 'mary pickford', 'between the sheets', 'faust', 'planter\'s punch', 'cuba libre',
    'rusty nail', 'remember the maine', 'manhattan', 'boulevardier', 'old fashioned', 'whiskey sour',
    'vieux carré', 'ward 8', 'mint julep', 'rob roy',
    'brandy crusta', 'sidecar', 'sazerac', 'stinger', 'alexander', 'porto flip', 'french connection',
    'margarita', 'black russian', 'metropolitan'
]

# 2. 필터링 (소문자 기준으로 검색)
filtered_df = clean_df[clean_df['strDrink'].str.lower().isin(cocktail_keywords)].copy()

# 3. DB에 넣을 포맷으로 변환
db_rows = []
for idx, row in filtered_df.iterrows():
    db_rows.append({
        'recipe_id': idx,  # 임시 고유 ID로 인덱스 사용
        'name': row['strDrink'],

        'ingredient1_id': row.get('strIngredient1'),
        'ingredient1_amount': row.get('strMeasure1'),
        'ingredient2_id': row.get('strIngredient2'),
        'ingredient2_amount': row.get('strMeasure2'),
        'ingredient3_id': row.get('strIngredient3'),
        'ingredient3_amount': row.get('strMeasure3'),
        'ingredient4_id': row.get('strIngredient4'),
        'ingredient4_amount': row.get('strMeasure4'),

        'method': None,
        'ice_in_shake': None,
        'is_on_the_rocks': None,
        'glass_type': None,
        'abv': None,

        'summary': None,
        'creator_id': None,
        'created_at': None
    })

# 4. 데이터프레임으로 변환
result_df = pd.DataFrame(db_rows)

# 5. 미리보기
result_df
```
