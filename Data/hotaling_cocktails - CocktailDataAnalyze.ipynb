```python
# 기주 키워드 매핑 (브랜드/스타일 포함)
spirit_keywords = {
    "whiskey": ["bourbon", "rye", "scotch", "whiskey", "nikka", "hibiki", "jameson", "irish whiskey"],
    "brandy": ["cognac", "armagnac", "h by hine"],
    "gin": ["no3 gin", "junipero", "london dry", "old tom", "pink gin", "plymouth", "gin"],
    "vodka": ["vodka", "karlssons gold vodka"],
    "rum": ["rum", "pink pigeon rum"],
    "mezcal": ["tequila", "mezcal"],
    "absinthe": ["absinthe"]
}

# 재료 표현 통일 매핑
ingredient_aliases = {
    "fresh lemon juice": "lemon juice",
    "lemon juice": "lemon juice",
    "fresh lime juice": "lime juice",
    "lime juice": "lime juice",
    "syrup": "simple syrup",
    "simple syrup": "simple syrup",
    "sugar syrup": "simple syrup",
    "triple sec": "orange liqueur",
    "cointreau": "orange liqueur",
    "orange liqueur": "orange liqueur",
    "sweet vermouth": "sweet vermouth",
    "dry vermouth": "dry vermouth",
    "pineapple juice": "pineapple juice",
    "cranberry juice": "cranberry juice",
    "grapefruit juice": "grapefruit juice"
}
```

```python
# ml → oz 단위 변환
def convert_ml_to_oz(ingredient):
    match = re.search(r"(\d*\.?\d+)\s*ml", ingredient)
    if match:
        ml = float(match.group(1))
        oz = round(ml / 29.5735, 2)
        return re.sub(r"\d*\.?\d+\s*ml", f"{oz} oz", ingredient)
    return ingredient

# 기주 키워드 정규화
def normalize_spirit_keywords(text):
    for spirit, keywords in spirit_keywords.items():
        for keyword in keywords:
            # 'junipero gin' 같은 조합을 우선 처리
            pattern = rf"\b{re.escape(keyword)}\s+{spirit}\b"
            text = re.sub(pattern, spirit, text, flags=re.IGNORECASE)
            # 단일 키워드도 대체
            if keyword != spirit:
                pattern_single = rf"\b{re.escape(keyword)}\b"
                text = re.sub(pattern_single, spirit, text, flags=re.IGNORECASE)
    return text

# 재료 이름 정규화
def normalize_ingredient_name(name):
    name = name.strip().lower()
    return ingredient_aliases.get(name, name)

# 전체 재료 문자열 정제
def clean_ingredient_string(ingredient_text):
    if pd.isna(ingredient_text):
        return None
    ingredients = ingredient_text.lower()
    ingredients = ",".join([convert_ml_to_oz(i.strip()) for i in ingredients.split(',')])
    ingredients = normalize_spirit_keywords(ingredients)
    return ingredients
```

```python
from collections import Counter
import pandas as pd
import re

# -----------------------------
# 1. 재료 테이블 기준 목록 + 기주 목록
# -----------------------------
ingredient_table = [
    "lemon juice", "lime juice", "orange liqueur", "orange juice", "cherry liqueur",
    "cranberry juice", "pineapple juice", "almond liqueur", "simple syrup",
    "coffee liqueur", "herbal liqueur", "dry vermouth", "sweet vermouth",
    "bitters", "milk", "cream", "egg white"
]

base_spirits = ["gin", "vodka", "whiskey", "rum", "brandy", "mezcal", "absinthe"]

full_ingredient_list = ingredient_table + base_spirits

# -----------------------------
# 2. 재료 이름 정규화 함수
# -----------------------------
def normalize_ingredient_name(name):
    name = name.strip().lower()

    # 기주 우선 처리
    for spirit in base_spirits:
        if spirit in name:
            return spirit
    
    # 재료 테이블 이름 포함 여부
    for ing in ingredient_table:
        if ing in name:
            return ing

    return name  # 기타는 그대로

# -----------------------------
# 3. 정제된 재료 리스트 만들기 (Ingredients_Clean 기준)
# -----------------------------
def extract_ingredient_names(ingredients_clean):
    if pd.isna(ingredients_clean):
        return []
    ingredients = ingredients_clean.split(',')
    result = []
    for item in ingredients:
        # 수량 제거
        name = re.sub(r"\d*\.?\d+\s*(oz|ml|dash)", "", item).strip()
        name = re.sub(r"[^\w\s]", "", name).strip()
        normalized = normalize_ingredient_name(name)
        result.append(normalized)
    return result

# -----------------------------
# 4. 전체 재료 리스트로부터 카운트 계산
# -----------------------------
df['ingredient_list'] = df['Ingredients_Clean'].apply(extract_ingredient_names)
flat_ingredients = [i for sublist in df['ingredient_list'] for i in sublist]
ingredient_counts = Counter(flat_ingredients)

# -----------------------------
# 5. 기주 및 재료 테이블 기준 필터링 + 정렬
# -----------------------------
df_filtered = pd.DataFrame(ingredient_counts.items(), columns=['ingredient', 'count'])
df_filtered = df_filtered[df_filtered['ingredient'].isin(full_ingredient_list)]
df_filtered = df_filtered.sort_values(by='count', ascending=False).reset_index(drop=True)

# -----------------------------
# 6. 결과 확인
# -----------------------------
df_filtered.head(20)
```

```python
import pandas as pd
import re

# 1. CSV 불러오기
df = pd.read_csv('hotaling_cocktails - Cocktails.csv')
```

```python
# 2. 기주 정규화 키워드 맵
spirit_keywords = {
    "whiskey": ["bourbon", "rye", "scotch", "whiskey", "nikka", "jameson", "hibiki", "irish whiskey"],
    "brandy": ["cognac", "armagnac"],
    "gin": ["gin", "juniper", "london dry", "old tom", "pink gin", "plymouth"],
    "vodka": ["vodka"],
    "rum": ["rum"],
    "mezcal": ["tequila", "mezcal"],
    "absinthe": ["absinthe"]
}

# 3. ml → oz 변환 함수
def convert_ml_to_oz(ingredient):
    match = re.search(r"(\d*\.?\d+)\s*ml", ingredient, re.IGNORECASE)
    if match:
        ml = float(match.group(1))
        oz = round(ml / 29.5735, 2)
        return re.sub(r"\d*\.?\d+\s*ml", f"{oz} oz", ingredient)
    return ingredient

# 4. 기주 키워드 정규화 함수
def normalize_spirit_keywords(text):
    for spirit, keywords in spirit_keywords.items():
        for keyword in keywords:
            # 예: 'junipero gin' → 'gin'
            pattern = rf"\b{re.escape(keyword)}\s+{spirit}\b"
            text = re.sub(pattern, spirit, text, flags=re.IGNORECASE)
            # 단독 키워드도 체크 (예: 'hibiki')
            if keyword != spirit:
                pattern_single = rf"\b{re.escape(keyword)}\b"
                text = re.sub(pattern_single, spirit, text, flags=re.IGNORECASE)
    return text

# 5. 전체 재료 정제 함수
def clean_ingredient_string(ingredient_text):
    if pd.isna(ingredient_text):
        return None
    ingredients = ingredient_text.lower()
    ingredients = ",".join([convert_ml_to_oz(i.strip()) for i in ingredients.split(',')])
    ingredients = normalize_spirit_keywords(ingredients)
    return ingredients

# 6. 전처리 적용
df['Ingredients_Clean'] = df['Ingredients'].apply(clean_ingredient_string)

# 7. 기주 추출 함수
base_spirits = ['gin', 'vodka', 'whiskey', 'rum', 'mezcal', 'brandy', 'absinthe']

def extract_base_spirit(ingredients_str):
    if pd.isna(ingredients_str):
        return None
    ingredients = ingredients_str.lower().split(',')
    max_amount = 0
    selected_base = None
    for item in ingredients:
        for base in base_spirits:
            if base in item:
                match = re.search(r"(\d*\.?\d+)\s*oz", item)
                if match:
                    amount = float(match.group(1))
                    if amount > max_amount:
                        max_amount = amount
                        selected_base = base
    return selected_base

# 8. 주재료 추출 함수 (기주 제외)
def extract_main_ingredient_excluding_base(ingredients_str, base_spirit):
    if pd.isna(ingredients_str):
        return None
    ingredients = ingredients_str.lower().split(',')
    max_amount = 0
    main_ingredient = None
    for item in ingredients:
        if base_spirit and base_spirit in item:
            continue
        match = re.search(r"(\d*\.?\d+)\s*oz", item)
        if match:
            amount = float(match.group(1))
            name = re.sub(r"\d*\.?\d+\s*oz", "", item).strip()
            name = re.sub(r"[^\w\s]", "", name).strip()
            if amount > max_amount:
                max_amount = amount
                main_ingredient = name
    return main_ingredient

# 9. 가니시 추출 함수
def extract_garnish_list(garnish_str):
    if pd.isna(garnish_str):
        return []
    return [g.strip().lower() for g in garnish_str.split(',')]
```

```python
# 다양한 표현을 통일할 재료 이름 매핑 사전
ingredient_aliases = {
    "fresh lemon juice": "lemon juice",
    "lemon juice": "lemon juice",
    "fresh lime juice": "lime juice",
    "lime juice": "lime juice",
    "simple syrup": "simple syrup",
    "sugar syrup": "simple syrup",
    "syrup": "simple syrup",
    "orange liqueur": "orange liqueur",
    "triple sec": "orange liqueur",
    "cointreau": "orange liqueur",
    "sweet vermouth": "sweet vermouth",
    "dry vermouth": "dry vermouth",
    "pineapple juice": "pineapple juice",
    "cranberry juice": "cranberry juice",
    "grapefruit juice": "grapefruit juice"
    # 필요 시 추가 가능
}
```

```python
def normalize_ingredient_name(name):
    name = name.strip().lower()
    return ingredient_aliases.get(name, name)  # 없으면 그대로 반환
```

```python
def extract_all_normalized_ingredients(ingredients_clean):
    if pd.isna(ingredients_clean):
        return []
    ingredients = ingredients_clean.split(',')
    cleaned = []
    for item in ingredients:
        # 용량 제거
        item = re.sub(r"\d*\.?\d+\s*oz", "", item).strip()
        item = re.sub(r"[^\w\s]", "", item).strip()
        norm = normalize_ingredient_name(item)
        cleaned.append(norm)
    return cleaned
```

```python
def extract_all_normalized_ingredients(ingredients_clean):
    if pd.isna(ingredients_clean):
        return []
    ingredients = ingredients_clean.split(',')
    cleaned = []
    for item in ingredients:
        # 용량 제거
        item = re.sub(r"\d*\.?\d+\s*oz", "", item).strip()
        item = re.sub(r"[^\w\s]", "", item).strip()
        norm = normalize_ingredient_name(item)
        cleaned.append(norm)
    return cleaned
```

```python
df['normalized_ingredients'] = df['Ingredients_Clean'].apply(extract_all_normalized_ingredients)
```

```python
# 10. 새로운 summary DataFrame 생성
summary_data = []

for _, row in df.iterrows():
    name = row['Cocktail Name']
    ingredients_clean = row['Ingredients_Clean']
    base = extract_base_spirit(ingredients_clean)
    main = extract_main_ingredient_excluding_base(ingredients_clean, base)
    garnishes = extract_garnish_list(row['Garnish'])

    summary_data.append({
        'cocktail_name': name,
        'base_spirit': base,
        'main_ingredient': main,
        'garnishes': garnishes
    })

df_summary = pd.DataFrame(summary_data)

# 11. 결과 확인
df_summary.head(10)
```
